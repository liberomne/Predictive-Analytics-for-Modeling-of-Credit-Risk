## Analisis Distribusi & Outlier

````python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

exclude_columns = ['id', 'member_id']
df_numeric = df.select_dtypes(include=['number']).drop(columns=exclude_columns, errors='ignore')
valid_columns = [col for col in df_numeric.columns if df_numeric[col].dropna().nunique() > 1]
num_cols = len(valid_columns)
num_rows = (num_cols // 4) + (1 if num_cols % 4 != 0 else 0)

plt.figure(figsize=(15, 5 * num_rows))
for i, column in enumerate(valid_columns):
    plt.subplot(num_rows, 4, i + 1)
    sns.boxplot(x=df_numeric[column].dropna())
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)

plt.tight_layout()
plt.show()
````

1. **Fitur dengan Banyak Outlier (Ekstrem Kanan)**
Contohnya:
* `loan_amnt`, `annual_inc`, `revol_bal`, `total_pymnt`, `recoveries`, dll.
* Sebagian besar data terkonsentrasi dalam kisaran kecil.
* Ada nilai-nilai yang jauh lebih tinggi dibanding mayoritas → **potensi outlier positif ekstrim**.
* Bisa disebabkan oleh user dengan penghasilan sangat tinggi, pinjaman besar, atau pembayaran besar.

2. **Distribusi Skewed (Condong)**
Fitur seperti `recoveries`, `collection_recovery_fee`, `last_pymnt_amnt`, dll menunjukkan distribusi **skewed right (condong ke kanan)**:
* Banyak data kecil mendekati nol
* Outlier berupa nilai besar
* Mayoritas orang memiliki jumlah recoveries kecil atau nol, namun sebagian kecil memiliki nilai yang besar.

3. **Fitur dengan Banyak Nilai Nol**
Beberapa boxplot menunjukkan **strip vertikal padat di nol**:
* Misalnya `collections_12_mths_ex_med`, `mths_since_last_major_derog`, dll.
* Banyak data bernilai nol → dominasi populasi tidak memiliki kejadian tersebut.
* Harus dipertimbangkan saat modeling: apakah kolom ini informatif atau tidak.

## Mengatasi Outlier

````python
main_df
````

````python
def remove_outliers_iqr(main_df, column):
    Q1 = main_df[column].quantile(0.25)
    Q3 = main_df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    filtered_main_df = main_df[(main_df[column] >= lower_bound) & (main_df[column] <= upper_bound)]
    print(f"{column}: {main_df.shape[0] - filtered_main_df.shape[0]} outliers removed.")
    return filtered_main_df
numerical_cols = ['loan_amnt', 'funded_amnt', 'int_rate', 'installment', 'total_pymnt', 'recoveries']

for col in numerical_cols:
    main_df = remove_outliers_iqr(main_df, col)
````

## Logistic Regression & Random Forest

````python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

logreg = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)
logreg.fit(X_train, y_train)
y_pred_logreg = logreg.predict(X_test)

rf = RandomForestClassifier(class_weight='balanced', n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
````

````python
print("=== Logistic Regression Report ===")
print(classification_report(y_test, y_pred_logreg))
print(confusion_matrix(y_test, y_pred_logreg))

print("\n=== Random Forest Report ===")
print(classification_report(y_test, y_pred_rf))
print(confusion_matrix(y_test, y_pred_rf))
````

## Evaluasi Kinerja Model 2

````python
from sklearn.metrics import classification_report, confusion_matrix
import pandas as pd

# Evaluasi Model
def evaluate_model(model, X_test, y_test, model_name="Model"):
    y_pred = model.predict(X_test)

    # Metrik Evaluasi
    report = classification_report(y_test, y_pred, output_dict=True)
    confusion = confusion_matrix(y_test, y_pred)

    # Ringkasan Evaluasi dalam bentuk DataFrame untuk lebih jelas
    summary = pd.DataFrame(report).transpose()
    summary["Model"] = model_name

    # Menampilkan Metrik
    print(f"=== {model_name} Report ===")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    print("\nConfusion Matrix:")
    print(confusion)

    return summary, confusion

# Model yang sudah dilatih sebelumnya (logreg, rf, gb, xgb, lgb) harus sudah tersedia
models = [logreg, rf, gb, xgb, lgb]
model_names = ["Logistic Regression", "Random Forest", "Gradient Boosting", "XGBoost", "LightGBM"]

# Menyimpan hasil evaluasi untuk semua model
evaluation_results = []

for model, name in zip(models, model_names):
    model_summary, confusion = evaluate_model(model, X_test, y_test, name)
    evaluation_results.append(model_summary)

# Menyusun ringkasan semua model ke dalam satu DataFrame
final_summary = pd.concat(evaluation_results, ignore_index=True)
print("\n=== Final Summary ===")
print(final_summary)
````
